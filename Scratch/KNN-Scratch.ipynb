```python
import random
import math
```

```python
def Euclidean_distance(x1,x2,y1,y2):
  return math.sqrt((x1 - x2)**2 + (y1-y2)**2)

def select_centroid(X,Y):
  c1_x = random.choice(X)
  c2_x = random.choice(X)
  c1_y = random.choice(Y)
  c2_y = random.choice(Y)

  return [(c1_x,c1_y),(c2_x,c2_y)]
```

```python
X = [1,3,5,7,2,4]
Y = [2,4,6,8,3,6]

centroid_1,centroid_2 = select_centroid(X,Y)

Cluster_1 = {"Centroid": (centroid_1[0],centroid_1[1]), "Points": []}
Cluster_2 = {"Centroid": (centroid_2[0],centroid_2[1]), "Points": []}

for i in range(len(X)):
  res_1 = Euclidean_distance(X[i],centroid_1[0],Y[i],centroid_1[1])
  res_2 = Euclidean_distance(X[i],centroid_2[0],Y[i],centroid_2[1])

  if res_1 <= res_2:
    Cluster_1["Points"].append((X[i],Y[i]))
  else:
    Cluster_2["Points"].append((X[i],Y[i]))

print("Cluster-1 : ",Cluster_1)
print("Cluster-2 : ",Cluster_2)
```
