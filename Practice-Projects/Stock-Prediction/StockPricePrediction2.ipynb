```python
import yfinance as yf
infy = yf.Ticker("infy.NS")
earning_history = infy.get_balance_sheet()
```

```python
data = infy.history("1y")
data.columns = data.columns.str.replace(' ','_')
data.reset_index(inplace=True)
data.head(10)
```

```python
def check_Stock_splits(data):
  if sum(data.Stock_Splits) < 0.0:
    data = data.drop(labels=["Stock_Splits"],inplace=True)
  return data
```

```python
data = check_Stock_splits(data)
```
