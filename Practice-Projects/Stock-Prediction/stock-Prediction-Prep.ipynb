```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Machine Learning /My-Projects/20-12-2023-TO-20-12-2024-SUZLON-ALL-N.csv")
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('â‚¹', 'INR')
data['turnover'] = pd.to_numeric(data['turnover_INR'].str.replace(',', ''), errors='coerce')
data['total_trades'] = pd.to_numeric(data['no._of_trades'].str.replace(',',''))
data.info()
```

```python
# moving trend analysis
data['MA_10'] = data['close_price'].rolling(window=10).mean()
print(data['MA_10'][249])
data['MA_50'] = data['close_price'].rolling(window=50).mean()
print(data['MA_50'][249])
```

```python
# daily volantility
data['Daily_Volatility'] = data['high_price'] - data['low_price']
print(data['Daily_Volatility'][249])
```

```python
#Turn over ratio
data['Turn_over_ratio'] = data['turnover'] / data['total_trades']
print(data['Turn_over_ratio'][249])
```

```python
# 7. RSI (Relative Strength Index)
def calculate_rsi(data, window=14):
    delta = data['close_price'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data['RSI'] = calculate_rsi(data)
print(data['RSI'][249])
```

```python
# Bollinger Bands
# 1. Get Moving average of 20 days
data['MA_20'] = data['close_price'].rolling(window=20).mean()
print(data['MA_20'][249])
# 2. Get Standard Deviation for 20 days
data['STD_20'] = data['close_price'].rolling(window=20).std()
print(data['STD_20'][249])
# Upper band
data['Upper_band'] = data['MA_20'] + (2 * data['STD_20'])
print(data['Upper_band'][249])
# lower band
data['lower_band'] = data['MA_20'] - (2 * data['STD_20'])
print(data['lower_band'][249])
```
