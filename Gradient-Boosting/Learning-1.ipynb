# Day-2 Model - Validation


```python
import pandas as pd
```

```python
DataFrame = pd.read_csv('/content/train.csv')
print(DataFrame.columns)
```

```python
# All the Necessary data needed
Feature_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = DataFrame[Feature_columns] # Features
Y = DataFrame.SalePrice # Prediction data
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
# Now split Training data and Validation Data
Val_X , Train_X , Val_Y , Train_Y = train_test_split(X,Y,random_state=1)
# Now Train Model using Training Data
House_Model = DecisionTreeRegressor(random_state = 1)
House_Model.fit(Train_X,Train_Y)
# Predition of Data
predicted_data = House_Model.predict(Val_X)
predicted_data
# finding the MAE of the data to know how much it is accurate
Error_in_data = mean_absolute_error(Val_Y,predicted_data)
print("Error in Predicted Data : ",Error_in_data)
```

```python
# without using Data split
House_Model.fit(X,Y)
predicted_data_2 = House_Model.predict(X)
print(predicted_data_2)
Error_in_data_2 = mean_absolute_error(Y,predicted_data_2)
print("Error in data 2 : ",Error_in_data_2)
```

**Conclusion here we can see if same data is used for training and validation the model becomes very inaccurate**
